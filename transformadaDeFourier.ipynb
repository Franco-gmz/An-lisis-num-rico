```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

#Ejercicio F4

##Funci√≥n

```python
t = np.linspace(0,np.pi,1000)
f = lambda t: 4*np.sin(2*t)+2*np.sin(6*t)
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(t,f(t))
ax.set_title("f(t)")
ax.grid()

plt.show()
```

##Transformada

```python
j = np.arange(0,10)
x = np.array([0.00000,4.25325,2.62866,2.62866,4.25325,0.00000,-4.25325,-2.62866,-2.62866,-4.25325])
T = (2*np.pi)/20
f = 1/T
```

```python
X = fft(x)
intensidad = np.abs(X)
frecuencias_angulares = (2*np.pi*f)*j/len(j)
```

```python
plt.figure(figsize=(8,5))
plt.plot(frecuencias_angulares,intensidad)

plt.axvline(2, 0, 20,color='r', label='w = 2')
plt.axvline(6, 0, 20,color='r', label='w = 6')
plt.legend()

plt.grid()
plt.xlabel("Frecuencia angular w")
plt.ylabel("Intensidad")
plt.show()
```
